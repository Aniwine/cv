```python
import numpy as np
import matplotlib.pyplot as plt


def create_grid(weight=100,height=100):
    mr = height
    mc = weight

    #xx是横向，yy是纵向
    xx = np.arange(mc-1,-1,-1)
    yy = np.arange(0,mr,1)
    #Y是纵向，X是横向
    [Y, X] = np.meshgrid(xx, yy)
    ms = np.transpose(np.asarray([X.flatten('F'), Y.flatten('F')]), (1,0))

    perturbed_mesh = ms
    """nv控制扭曲次数"""
    nv = np.random.randint(40) - 1
    for k in range(nv):
        #Choosing one vertex randomly
        """vidx控制扭曲点位置"""
        vidx = np.random.randint(1,np.shape(ms)[0])
        # vtex = ms[vidx, :]
        vtex=ms[300,:]
        #Vector between all vertices and the selected one
        xv  = perturbed_mesh - vtex
        #Random movement 
        # mv = (np.random.rand(1,2) - 0.5)*20
        # mv=(np.random.rand(1,2) - 0.5)*20-vtex
        """mv控制沿着哪一根向量扭曲"""
        mv=np.array([0,9])
        hxv = np.zeros((np.shape(xv)[0], np.shape(xv)[1] +1) )
        hxv[:, :-1] = xv
        hmv = np.tile(np.append(mv, 0), (np.shape(xv)[0],1))
        d = np.cross(hxv, hmv)
        d = np.absolute(d[:, 2])
        d = d / (np.linalg.norm(mv, ord=2))
        wt = d
        
        curve_type = np.random.rand(1)
        if curve_type > 2:
            alpha = np.random.rand(1) * 50 + 50
            wt = alpha / (wt + alpha)
        else:
            """alpha控制扭曲程度"""
            # alpha = np.random.rand(1) + 1
            alpha=1.2
            # alpha=0.01

            #这里除以100是归一化d
            wt = 1 - (wt / 1280 )**alpha
        #wv
        msmv = mv * np.expand_dims(wt, axis=1)
        #p+wv
        perturbed_mesh = perturbed_mesh + msmv

    plt.scatter(perturbed_mesh[:, 0], perturbed_mesh[:, 1], c=np.arange(0, mr*mc))
    plt.show()

    #边界检查
    perturbed_mesh[:, 0] = np.where(perturbed_mesh[:, 0] > weight, weight, perturbed_mesh[:, 0])
    perturbed_mesh[:, 1] = np.where(perturbed_mesh[:, 1] > height, height, perturbed_mesh[:, 1])

    return perturbed_mesh[:,0],perturbed_mesh[:,1]
```

```python
import cv2

src_img = cv2.imread('/volume/Surface-flatting/DocUNet/data/edu_photo/29.jpg')
height, width, _ = src_img.shape
dh = height // 10
dw = width // 10
img = cv2.copyMakeBorder(src_img, dh, dh, dw, dw, borderType=cv2.BORDER_CONSTANT,value=(0,0,0))
nh, nw, _ = img.shape
xs, ys  = create_grid(weight=nw, height=nh)
xs = xs.reshape(nh, nw).astype(np.float32)
ys = ys.reshape(nh, nw).astype(np.float32)
dst = cv2.remap(src_img, xs, ys, cv2.INTER_CUBIC)
dst=cv2.flip(dst,0)
cv2.imwrite('/volume/Surface-flatting/DocUNet/data/edu_photo/29_result.jpg', dst)
```

```python
#场景不需要使用折叠，只考虑弯曲，且弯曲一般为平行于书本分界线的直线
import numpy as np
import matplotlib.pyplot as plt

mr = 88
mc = 68

xx = np.arange(mr-1, -1, -1)
yy = np.arange(0, mc, 1)
[Y, X] = np.meshgrid(xx, yy)
ms = np.transpose(np.asarray([X.flatten('F'), Y.flatten('F')]), (1,0))

perturbed_mesh = ms
nv = np.random.randint(20) - 1
for k in range(nv):
    #Choosing one vertex randomly
    vidx = np.random.randint(np.shape(ms)[0])
    vtex = ms[vidx, :]
    #Vector between all vertices and the selected one
    xv  = perturbed_mesh - vtex
    #Random movement
    """
    改进点
    确定向量为一条平行于书本分界线的直线
    """ 
    # mv = (np.random.rand(1,2) - 0.5)*20
    mv=np.array([vtex[0],np.random.randint(0,10)])-vtex
    # mv=np.array([np.random.randint(0,10),vtex[1]])-vtex
    hxv = np.zeros((np.shape(xv)[0], np.shape(xv)[1] +1) )
    hxv[:, :-1] = xv
    hmv = np.tile(np.append(mv, 0), (np.shape(xv)[0],1))
    d = np.cross(hxv, hmv)
    d = np.absolute(d[:, 2])
    d = d / (np.linalg.norm(mv, ord=2))
    wt = d
    
    curve_type = np.random.rand(1)
    if curve_type > 0.3:
        alpha = np.random.rand(1) * 50 + 50
        wt = alpha / (wt + alpha)
    else:
        alpha = np.random.rand(1) + 1
        wt = 1 - (wt / 100 )**alpha
    msmv = mv * np.expand_dims(wt, axis=1)
    perturbed_mesh = perturbed_mesh + msmv

plt.scatter(perturbed_mesh[:, 0], perturbed_mesh[:, 1], c=np.arange(0, mr*mc))
plt.show()
```
